```python
cnt=0 
data_list = []
# len(list_total_count)
for idx in tqdm(range(1, list_total_count)):
    try:
        url ='http://openapi.seoul.go.kr:8088/{apiKey}/json/tbCycleUseStatus/{idx}/{idx}/{STAT_DATA}'.format(apiKey=apiKey,
                                                                                                STAT_DATA=STAT_DATA,idx=idx)
        response = requests.get(url)

        api = response.text
        dict = json.loads(api)
        data = dict.get('useStatus').get('row')
        data_list.append(data[0])

    except:
        cnt +=1
        if cnt%50 ==0:
            print(cnt,'안대는중')
        pass

data_df = pd.DataFrame(data_list)
data_df

    
```
