```python
import polars as pl
import geopy.distance
import pandas as pd

file_path = '/Users/iseunghyeon/Desktop/프로젝트/data/천안시_proj_data/버스정류장/Bus_around_500_08231838.pickle'
pandas_df = pd.read_pickle(file_path)
pandas_df['location']= pd.Series(zip(pandas_df['위도'],pandas_df['경도']))
bus_station = pl.DataFrame(pandas_df)

# Create a new column 'location' as a tuple of (latitude, longitude)
bus_station = bus_station.with_columns([
    # ('location', [tuple(lat, lon) for lat, lon in zip(bus_station['위도'], bus_station['경도'])]),
    pl.col('정류장번호'),   # Copy the existing column
    pl.col('정류장명')      # Copy the existing column
])

bus_station = bus_station.head(5)

# Function to calculate distance
def calculate_distance(bus_cp, company_house):
    data_list = []

    for idx in range(len(bus_cp)):
        for jdx in range(len(company_house)):
            distance = geopy.distance.distance(bus_cp[idx]['location'], company_house[jdx]['location']).km

            data = {'정류장번호': bus_cp[idx]['정류장번호'], '정류장명': bus_cp[idx]['정류장명'], 'location': bus_cp[idx]['location'],
                    '정류장번호2': company_house[jdx]['정류장번호'], '정류장명2': company_house[jdx]['정류장명'], 'distance': distance}

            data_list.append(data)

    data_df = pl.DataFrame(data_list)
    return data_df

bus_target_distance = calculate_distance(bus_station, bus_station)
pandas_df = bus_target_distance.to_pandas()
pandas_df['정류장번호'] = pandas_df['정류장번호'].apply(lambda elm : elm[0])
pandas_df['정류장번호2'] = pandas_df['정류장번호2'].apply(lambda elm : elm[0])

pandas_df
```
